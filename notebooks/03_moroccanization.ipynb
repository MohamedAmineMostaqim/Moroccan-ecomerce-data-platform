# Moroccanization of the Olist E-Commerce Dataset

## Objective

The original Olist dataset is based on the Brazilian e-commerce market.

To create a realistic Moroccan business intelligence project, the dataset was transformed by replacing Brazilian-specific values with Moroccan equivalents while preserving the overall structure and analytical integrity of the data.

This notebook validates the transformation by comparing the original and Moroccanized datasets.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RAW = Path("../data/raw")
PROCESSED = Path("../data/processed")
```

```python
customers_raw = pd.read_csv(RAW / "olist_customers_dataset.csv")
customers_ma = pd.read_csv(PROCESSED / "customers.csv")

sellers_raw = pd.read_csv(RAW / "olist_sellers_dataset.csv")
sellers_ma = pd.read_csv(PROCESSED / "sellers.csv")

payments_raw = pd.read_csv(RAW / "olist_order_payments_dataset.csv")
payments_ma = pd.read_csv(PROCESSED / "payments.csv")

products_raw = pd.read_csv(RAW / "olist_products_dataset.csv")
products_ma = pd.read_csv(PROCESSED / "products.csv")
```

```python
comparison = pd.DataFrame({
    "Dataset": [
        "Customers",
        "Sellers",
        "Products",
        "Payments"
    ],
    "Raw Rows": [
        len(customers_raw),
        len(sellers_raw),
        len(products_raw),
        len(payments_raw)
    ],
    "Processed Rows": [
        len(customers_ma),
        len(sellers_ma),
        len(products_ma),
        len(payments_ma)
    ]
})

comparison
```

# Missing Values Comparison

```python
for name, raw_df, ma_df in [
    ("Customers", customers_raw, customers_ma),
    ("Sellers", sellers_raw, sellers_ma),
    ("Products", products_raw, products_ma),
    ("Payments", payments_raw, payments_ma),
]:
    print("=" * 50)
    print(name)

    print("\nRaw")
    print(raw_df.isna().sum())

    print("\nProcessed")
    print(ma_df.isna().sum())

    print()
```

# Customer Cities

```python
fig, ax = plt.subplots(figsize=(12,5))

customers_raw["customer_city"] \
.value_counts() \
.head(10) \
.plot(kind="bar", ax=ax)

ax.set_title("Top Brazilian Cities")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))

customers_ma["customer_city"] \
.value_counts() \
.head(10) \
.plot(kind="bar", ax=ax)

ax.set_title("Top Moroccan Cities")

plt.show()
```
